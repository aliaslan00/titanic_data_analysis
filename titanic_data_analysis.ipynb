```python
def degiskenler(variable):
    cat = train_data[variable]  # Belirli bir değişkenin kategorik değerlerini içeren bir veri serisi oluşturuluyor.
    sayi = cat.value_counts()  # Kategorik değerlerin sayıları hesaplanıyor.
    plt.figure(figsize=(9, 3))  # Bir grafik figürü oluşturuluyor ve boyutu ayarlanıyor.
    plt.bar(sayi.index, sayi)  # Çubuk grafik oluşturuluyor.
    plt.xticks(sayi.index, sayi.index.values)  # x eksenindeki etiketler ayarlanıyor.
    plt.ylabel(variable)  # y ekseninin etiketi ayarlanıyor.
    plt.show()  # Grafik ekrana çizdiriliyor.
    print(variable, sayi)  # Değişken adı ve kategorik değerlerin sayıları ekrana yazdırılıyor.
```

```python
category1 = ["Survived" , "Sex" , "Pclass" , "Embarked" , "SibSp" , "Parch"]
```

```python
for i in category1:
  degiskenler(i)
```

```python
category2 = ["Cabin" ,  "Name" , "Ticket"]
for i in category2:
  print(train_data[i].value_counts)
```

```python
def hist(variable):
  plt.figure(figsize = (9,3))
  plt.hist(train_data[variable])
  plt.xlabel(variable)
  plt.ylabel("frekans")
  plt.title(variable)
  plt.show
```

```python
sayisal = ["Fare" , "Age" , "PassengerId"]
for i in sayisal:
  hist(i)
```
